## Generating and Visualizing a Haiku

Come up with a HAIKU or ASCII art using SingleStore Notebooks. Below are some examples on creating Haiku for your reference.

This below code consists of two main parts:

1. Generating a Haiku: The generate_haiku function creates a simple haiku using pre-defined lists of phrases that correspond to the traditional 5-7-5 syllable structure of haikus. This function randomly selects one phrase from each list to construct the haiku.

2. Visualizing the Haiku: The visualize_haiku function uses matplotlib to create a visualization of the generated haiku. It sets up a figure with a custom background color, hides the axes for a cleaner look, and displays the haiku text in the center with a styled bounding box around it.

### Example 1

```python
import matplotlib.pyplot as plt
import numpy as np

# Simple Haiku Generator Function
def generate_haiku():
    # Example words for simplicity, replace or expand as needed
    five_syllables = ["The silent sunset", "Mountains in the fall", "Waves crash on the shore"]
    seven_syllables = ["Clouds cover the moon tonight", "A quiet pond reflects stars", "Leaves rustle in the wind's song"]
    
    # Randomly select lines to form a haiku
    haiku = [
        np.random.choice(five_syllables),
        np.random.choice(seven_syllables),
        np.random.choice(five_syllables)
    ]
    return haiku

# Advanced Visualization Function with matplotlib
def visualize_haiku(haiku):
    # Create a figure and axis for drawing
    fig, ax = plt.subplots(figsize=(8, 6))
    
    # Hide axes
    ax.axis('off')
    
    # Set background (using an example pattern or image)
    # For a more advanced approach, you could dynamically load an image that matches the theme of the haiku
    bg_color = '#f0eade'  # Example: a paper-like background color
    fig.patch.set_facecolor(bg_color)
    ax.set_facecolor(bg_color)
    
    # Display the haiku text
    haiku_text = "\n".join(haiku)
    bbox_props = dict(boxstyle="round,pad=0.3", ec="black", lw=2, facecolor="white")
    ax.text(0.5, 0.5, haiku_text, fontsize=20, va='center', ha='center', bbox=bbox_props)
    
    plt.show()

# Generate and Visualize Haiku
haiku = generate_haiku()
visualize_haiku(haiku)
```

### Example 2

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
from PIL import Image
import os

# Expanded Haiku Generator Function
def generate_haiku():
    five_syllables = ["The silent sunset", "Mountains in the fall", "Waves crash on the shore",
                      "Early morning dew", "Whispers of the night"]
    seven_syllables = ["Clouds cover the moon tonight", "A quiet pond reflects stars",
                       "Leaves rustle in the wind's song", "Footprints vanish in the snow",
                       "The city sleeps under stars"]
    
    haiku = [
        np.random.choice(five_syllables),
        np.random.choice(seven_syllables),
        np.random.choice(five_syllables)
    ]
    return haiku

# Advanced Visualization Function with Dynamic Background
def visualize_haiku(haiku):
    fig, ax = plt.subplots(figsize=(10, 8))

    ax.axis('off')

    # Example backgrounds
    backgrounds = {
        "The silent sunset": "sunset.jpg",
        "Mountains in the fall": "mountains.jpg",
        "Waves crash on the shore": "waves.jpg",
        # Add paths for new images here
    }
    
    haiku_theme = haiku[0]  # Use the first line to determine the theme
    bg_image_path = backgrounds.get(haiku_theme, "default_background.jpg")  # Fallback to a default image
    
    try:
        img = Image.open(bg_image_path)
        img = img.resize((fig.bbox.xmax, fig.bbox.ymax), Image.ANTIALIAS)
        ax.imshow(img, extent=[0, 1, 0, 1], aspect='auto')
    except FileNotFoundError:
        print(f"Background image {bg_image_path} not found. Using plain background.")
        bg_color = '#f0eade'
        fig.patch.set_facecolor(bg_color)
        ax.set_facecolor(bg_color)

    haiku_text = "\n".join(haiku)
    bbox_props = dict(boxstyle="round,pad=0.3", ec="black", lw=2, facecolor="white", alpha=0.7)
    ax.text(0.5, 0.5, haiku_text, fontsize=20, va='center', ha='center', bbox=bbox_props, wrap=True)
    
    plt.title("A Haiku for You", fontsize=24, weight='bold', color='black')
    
    plt.show()

# Generate and Visualize Haiku
haiku = generate_haiku()
visualize_haiku(haiku)
```

### Example 3

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches

# Haiku Generator with Environmental Themes
def generate_environmental_haiku():
    lines = {
        "start": ["Dawn's light on leaves", "Moonlit whispering pines", "Sunset's final glow"],
        "middle": ["Rivers carve the earth", "Mountains touch the sky's edge", "Forests breathe with life"],
        "end": ["Night's curtain falls", "Autumn's crisp exhale", "Spring's chorus awakens"]
    }
    
    haiku = [
        np.random.choice(lines["start"]),
        np.random.choice(lines["middle"]),
        np.random.choice(lines["end"])
    ]
    return haiku

# Dynamic Visualization Function
def visualize_dynamic_haiku(haiku):
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    ax.axis('off')
    
    # Background Color
    ax.set_facecolor("#e0f7fa")
    
    # Dynamic elements based on the Haiku
    if "Dawn's light" in haiku[0]:
        ax.add_patch(patches.Circle((5, 5), 3, color="#fdd835"))  # Sun
    elif "Moonlit" in haiku[0]:
        ax.add_patch(patches.Circle((5, 5), 3, color="#cfd8dc"))  # Moon
    elif "Sunset's" in haiku[0]:
        ax.add_patch(patches.Circle((5, 5), 3, color="#ef5350"))  # Setting Sun
    
    if "Rivers" in haiku[1]:
        ax.plot([2, 8], [2, 2], color="#29b6f6", linewidth=5)  # River
    elif "Mountains" in haiku[1]:
        ax.add_patch(patches.Polygon([[3,3], [5,7], [7,3]], closed=True, color="#8d6e63"))  # Mountain
    elif "Forests" in haiku[1]:
        ax.add_patch(patches.Rectangle((3, 2), 4, 4, color="#66bb6a"))  # Forest
    
    # Haiku Text
    ax.text(5, 1, "\n".join(haiku), ha='center', fontsize=12, style='italic', bbox=dict(facecolor='white', alpha=0.5))
    
    plt.show()

# Generate and Visualize Haiku
haiku = generate_environmental_haiku()
visualize_dynamic_haiku(haiku)
```

### Example 4

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

# Seasonal Transition Haiku Generator
def generate_seasonal_transition_haiku():
    seasons = ["Winter", "Spring", "Summer", "Fall"]
    transitions = {
        ("Winter", "Spring"): ["Snow melts, streams flow", "Green buds break through frost", "Life awakens anew"],
        ("Spring", "Summer"): ["Flowers bloom in warmth", "Longer days, brighter sun", "Nature in full swing"],
        ("Summer", "Fall"): ["Leaves turn gold and red", "Harvest moon rises high", "Chill whispers in the wind"],
        ("Fall", "Winter"): ["First frost coats the ground", "Days shorten, nights grow long", "Silence under the snow"],
    }
    
    # Randomly select a transition
    start_season = np.random.choice(seasons)
    end_season = seasons[(seasons.index(start_season) + 1) % len(seasons)]
    
    haiku_lines = transitions[(start_season, end_season)]
    
    return haiku_lines, start_season, end_season

# Visualization with Gradient
def visualize_seasonal_transition(haiku, start_season, end_season):
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.axis('off')
    
    # Seasonal Colors
    colors = {
        "Winter": "#ffffff",
        "Spring": "#4caf50",
        "Summer": "#ffa726",
        "Fall": "#fb8c00"
    }
    
    # Create Gradient
    cmap = LinearSegmentedColormap.from_list("season_transition", [colors[start_season], colors[end_season]])
    gradient = np.linspace(0, 1, 256)
    gradient = np.vstack((gradient, gradient))
    
    ax.imshow(gradient, aspect='auto', cmap=cmap, extent=[0, 10, 0, 10])
    
    # Display Haiku
    ax.text(5, 5, "\n".join(haiku), ha='center', va='center', fontsize=14, color="black")
    
    plt.show()

# Generate and Visualize
haiku, start_season, end_season = generate_seasonal_transition_haiku()
visualize_seasonal_transition(haiku, start_season, end_season)
```

### Example 5

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Day and Night Haiku Generator
def generate_day_night_haiku():
    day_lines = ["Sunrise ignites skies", "Noon's bright, unyielding gaze", "Sunset's soft goodbye"]
    night_lines = ["Moon's gentle glow", "Stars whisper ancient tales", "Night's deep, serene veil"]
    
    # Creating a Haiku that represents a cycle from day to night and back
    haiku = [np.random.choice(day_lines), np.random.choice(night_lines), np.random.choice(day_lines)]
    return haiku

# Circular Visualization for Day and Night Cycle
def visualize_day_night_cycle(haiku):
    fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'aspect': 'equal'})
    ax.axis('off')
    
    # Drawing the Sun and Moon
    sun = patches.Circle((0.3, 0.75), 0.2, color="#FFD54F", transform=ax.transAxes)  # Sun
    moon = patches.Circle((0.7, 0.25), 0.15, color="#CFD8DC", transform=ax.transAxes)  # Moon
    
    ax.add_patch(sun)
    ax.add_patch(moon)
    
    # Drawing the path representing the cycle
    cycle_path = patches.Arc((0.5, 0.5), 0.8, 0.8, angle=0, theta1=0, theta2=360, edgecolor='gray', transform=ax.transAxes)
    ax.add_patch(cycle_path)
    
    # Displaying the Haiku text in the center
    ax.text(0.5, 0.5, "\n".join(haiku), ha='center', va='center', fontsize=14, color="black")

    plt.show()

# Generate and Visualize the Day and Night Cycle Haiku
haiku = generate_day_night_haiku()
visualize_day_night_cycle(haiku)
```

### Example 6

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Ellipse, Polygon
from matplotlib.colors import LinearSegmentedColormap

# Enhanced Haiku Function
def generate_abstract_haiku():
    times_of_day = ["Morning", "Noon", "Evening", "Night"]
    elements = ["River", "Mountain", "Tree", "Stars"]
    moods = ["Serene", "Vibrant", "Mysterious", "Peaceful"]
    
    haiku = [
        f"{np.random.choice(times_of_day)}'s {np.random.choice(moods)} Light",
        f"{np.random.choice(elements)} in the {np.random.choice(moods)}",
        f"Echoes of {np.random.choice(elements)}"
    ]
    return haiku

# Visualization with Improved Text Integration
def visualize_abstract_landscape(haiku):
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.axis('off')
    
    # Determine the sky's gradient based on the time of day in the Haiku
    sky_colors = {
        "Morning": ["#FFDAB9", "#00BFFF"],
        "Noon": ["#87CEEB", "#00BFFF"],
        "Evening": ["#FFA07A", "#20B2AA"],
        "Night": ["#2F4F4F", "#000000"]
    }
    time_of_day = haiku[0].split("'")[0]
    sky_color = sky_colors.get(time_of_day, ["#87CEEB", "#00BFFF"])
    ax.imshow(np.linspace(0, 1, 256).reshape(-1, 1), aspect='auto', extent=[0, 10, 0, 10], cmap=LinearSegmentedColormap.from_list("sky", sky_color))
    
    # Simplified Examples of Environmental Elements
    if "River" in haiku[1]:
        river = Ellipse((5, 2), 9, 1, color="#1E90FF", alpha=0.7)
        ax.add_patch(river)
    if "Mountain" in haiku[1]:
        mountain = Polygon([[3,3], [5,7], [7,3]], closed=True, color="#A52A2A", alpha=0.7)
        ax.add_patch(mountain)
    if "Tree" in haiku[1]:
        tree = Ellipse((5, 5), 3, 6, color="#228B22", alpha=0.7)
        ax.add_patch(tree)
    if "Stars" in haiku[2]:
        for _ in range(50):
            x, y = np.random.rand(2) * 10
            star = Ellipse((x, y), 0.1, 0.1, color="#FFFFFF", alpha=0.8)
            ax.add_patch(star)
    
    # Improved Haiku Display in the Sky
    text_y_position = 9  # Positioning the text higher in the visualization
    for line in haiku:
        ax.text(5, text_y_position, line, ha='center', va='center', fontsize=12, color="white")
        text_y_position -= 1  # Moving each line down

    plt.show()

# Generate an abstract Haiku
haiku = generate_abstract_haiku()

# Visualize the Haiku with an enhanced abstract landscape
visualize_abstract_landscape(haiku)
```

```python
import requests
from PIL import Image
from io import BytesIO

# Define the ASCII characters that will be used to replace the pixels.
ASCII_CHARS = ["@", "#", "S", "%", "?", "*", "+", ";", ":", ",", "."]

def resize_image(image, new_width=100):
    width, height = image.size
    ratio = height / width / 1.65  # Correct for aspect ratio
    new_height = int(new_width * ratio)
    resized_image = image.resize((new_width, new_height))
    return resized_image

def grayify(image):
    grayscale_image = image.convert("L")
    return grayscale_image

def pixels_to_ascii(image):
    pixels = image.getdata()
    ascii_str = ''
    for pixel in pixels:
        ascii_str += ASCII_CHARS[pixel // 25]  # Map the pixel value to ASCII_CHARS
    return ascii_str

def process_image_from_url(image_url, new_width=100):
    # Fetch the image from the URL
    response = requests.get(image_url)
    if response.status_code == 200:
        # Open the image from the bytes in response content
        image = Image.open(BytesIO(response.content))
        
        # Process the image
        image = resize_image(image, new_width)
        image = grayify(image)
        ascii_str = pixels_to_ascii(image)

        # Format the ASCII string so that each line has `new_width` characters.
        img_width = image.width
        ascii_str_len = len(ascii_str)
        ascii_img = ""
        for i in range(0, ascii_str_len, img_width):
            ascii_img += ascii_str[i:i+img_width] + "\n"
        
        # Print the ASCII art
        print(ascii_img)
    else:
        print("Failed to retrieve the image from the URL")

# Example usage with a public image URL
image_url = 'https://i2-prod.mirror.co.uk/incoming/article7813478.ece/ALTERNATES/s1200d/William-Shakespeare.jpg'  # Replace with your image's URL
process_image_from_url(image_url, 100)
```
